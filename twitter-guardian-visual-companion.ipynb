# Twitter Guardian - Visual Design Companion

Interactive design workspace for building a free Twitter spam detection system.

**Key Goal**: 24/7 automated spam filtering without API costs

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import Rectangle, FancyBboxPatch
import numpy as np

# Set up dark theme to match TTR branding
plt.style.use('dark_background')

# TTR Brand Colors
ttr_yellow = '#FFCC00'
obsidian = '#141519'
foreground = '#ECE9E1'
navy = '#2F3E56'
taupe = '#A7A39A'
```

## System Architecture Options

Let's visualize different approaches for your free Twitter Guardian:

```python
# Architecture Comparison Visualization
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 8))
fig.patch.set_facecolor(obsidian)

# Option 1: Browser Automation
ax1.set_title('Option 1: Browser + Local AI\n(Recommended - $0 cost)', 
             color=ttr_yellow, fontsize=12, fontweight='bold', pad=20)

# Draw browser automation flow
boxes1 = [
    ('Your Computer', 0.1, 0.8, 0.8, 0.15),
    ('Playwright\n(Browser Automation)', 0.15, 0.6, 0.7, 0.12),
    ('Local AI\n(Ollama Llama 3.2)', 0.15, 0.4, 0.7, 0.12),
    ('Actions\n(Block/Mute)', 0.15, 0.2, 0.7, 0.12)
]

for text, x, y, w, h in boxes1:
    rect = FancyBboxPatch((x, y), w, h, 
                         boxstyle="round,pad=0.02",
                         facecolor=navy, 
                         edgecolor=foreground,
                         linewidth=1)
    ax1.add_patch(rect)
    ax1.text(x + w/2, y + h/2, text, 
            ha='center', va='center', 
            color=foreground, fontsize=9, fontweight='medium')

# Add arrows
ax1.annotate('', xy=(0.5, 0.6), xytext=(0.5, 0.8),
            arrowprops=dict(arrowstyle='->', color=ttr_yellow, lw=2))
ax1.annotate('', xy=(0.5, 0.4), xytext=(0.5, 0.6),
            arrowprops=dict(arrowstyle='->', color=ttr_yellow, lw=2))
ax1.annotate('', xy=(0.5, 0.2), xytext=(0.5, 0.4),
            arrowprops=dict(arrowstyle='->', color=ttr_yellow, lw=2))

ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)
ax1.axis('off')

# Option 2: API + Cloud
ax2.set_title('Option 2: Twitter API + Cloud\n(Expensive - $100+/month)', 
             color=taupe, fontsize=12, fontweight='bold', pad=20)

boxes2 = [
    ('Twitter API\n($100/month)', 0.1, 0.8, 0.8, 0.15),
    ('Cloud AI\n(Claude/GPT costs)', 0.15, 0.6, 0.7, 0.12),
    ('VPS/Cloud\n($20-40/month)', 0.15, 0.4, 0.7, 0.12),
    ('Actions via API', 0.15, 0.2, 0.7, 0.12)
]

for text, x, y, w, h in boxes2:
    rect = FancyBboxPatch((x, y), w, h, 
                         boxstyle="round,pad=0.02",
                         facecolor='#4a1a1a',  # Reddish for expensive
                         edgecolor=taupe,
                         linewidth=1)
    ax2.add_patch(rect)
    ax2.text(x + w/2, y + h/2, text, 
            ha='center', va='center', 
            color=taupe, fontsize=9)

ax2.set_xlim(0, 1)
ax2.set_ylim(0, 1)
ax2.axis('off')

# Option 3: Hybrid Approach
ax3.set_title('Option 3: Hybrid\n(Complex setup)', 
             color=taupe, fontsize=12, fontweight='bold', pad=20)

boxes3 = [
    ('Browser (Read)', 0.1, 0.8, 0.35, 0.15),
    ('API (Write)', 0.55, 0.8, 0.35, 0.15),
    ('Local + Cloud AI', 0.15, 0.6, 0.7, 0.12),
    ('Complex Logic', 0.15, 0.4, 0.7, 0.12)
]

for text, x, y, w, h in boxes3:
    rect = FancyBboxPatch((x, y), w, h, 
                         boxstyle="round,pad=0.02",
                         facecolor='#2a2a4a',  # Bluish for complex
                         edgecolor=taupe,
                         linewidth=1)
    ax3.add_patch(rect)
    ax3.text(x + w/2, y + h/2, text, 
            ha='center', va='center', 
            color=taupe, fontsize=9)

ax3.set_xlim(0, 1)
ax3.set_ylim(0, 1)
ax3.axis('off')

plt.tight_layout()
plt.show()
```

## Filtering Strategy Visualization

Based on your successful techsnif signal detection system, let's design a similar multi-tier approach:

```python
# Filtering Pipeline Visualization
fig, ax = plt.subplots(figsize=(16, 10))
fig.patch.set_facecolor(obsidian)

# Title
ax.text(0.5, 0.95, 'Twitter Guardian: 3-Layer Defense System', 
        ha='center', va='center', color=ttr_yellow, fontsize=16, fontweight='bold')

# Layer 1: Heuristic Filter
layer1 = FancyBboxPatch((0.05, 0.75), 0.9, 0.15,
                        boxstyle="round,pad=0.02",
                        facecolor=navy,
                        edgecolor=ttr_yellow,
                        linewidth=2)
ax.add_patch(layer1)

ax.text(0.1, 0.85, 'LAYER 1: Heuristic Filter (Zero Cost)', 
        color=ttr_yellow, fontsize=12, fontweight='bold')
ax.text(0.1, 0.81, '• Default avatar detection (image hash)', color=foreground, fontsize=10)
ax.text(0.1, 0.78, '• Username patterns (random_12345678)', color=foreground, fontsize=10)
ax.text(0.5, 0.81, '• Bio keywords (crypto scam, "DM for deals")', color=foreground, fontsize=10)
ax.text(0.5, 0.78, '• Follower/following ratio anomalies', color=foreground, fontsize=10)

ax.text(0.85, 0.82, '80% blocked\n$0 cost', ha='center', va='center', 
        color=ttr_yellow, fontsize=11, fontweight='bold',
        bbox=dict(boxstyle="round,pad=0.3", facecolor=obsidian, edgecolor=ttr_yellow))

# Layer 2: Embedding Similarity  
layer2 = FancyBboxPatch((0.05, 0.55), 0.9, 0.15,
                        boxstyle="round,pad=0.02",
                        facecolor='#2E9B6E',  # Green accent
                        edgecolor=foreground,
                        linewidth=2)
ax.add_patch(layer2)

ax.text(0.1, 0.65, 'LAYER 2: Embedding Similarity (Local)', 
        color=foreground, fontsize=12, fontweight='bold')
ax.text(0.1, 0.61, '• Pre-computed spam profile embeddings', color=foreground, fontsize=10)
ax.text(0.1, 0.58, '• Sentence-transformers (runs locally)', color=foreground, fontsize=10)
ax.text(0.5, 0.61, '• Cosine similarity matching', color=foreground, fontsize=10)
ax.text(0.5, 0.58, '• Learning from blocked patterns', color=foreground, fontsize=10)

ax.text(0.85, 0.62, '15% more\n~$0 cost', ha='center', va='center', 
        color=foreground, fontsize=11, fontweight='bold',
        bbox=dict(boxstyle="round,pad=0.3", facecolor=obsidian, edgecolor=foreground))

# Layer 3: LLM Verification
layer3 = FancyBboxPatch((0.05, 0.35), 0.9, 0.15,
                        boxstyle="round,pad=0.02",
                        facecolor='#7C3AED',  # Purple accent
                        edgecolor=foreground,
                        linewidth=2)
ax.add_patch(layer3)

ax.text(0.1, 0.45, 'LAYER 3: LLM Verification (Rare)', 
        color=foreground, fontsize=12, fontweight='bold')
ax.text(0.1, 0.41, '• Local Ollama Llama 3.2 3B model', color=foreground, fontsize=10)
ax.text(0.1, 0.38, '• Only for sophisticated edge cases', color=foreground, fontsize=10)
ax.text(0.5, 0.41, '• "Is this genuine tech professional?"', color=foreground, fontsize=10)
ax.text(0.5, 0.38, '• Appeals and uncertain cases', color=foreground, fontsize=10)

ax.text(0.85, 0.42, '<5% traffic\nMinimal cost', ha='center', va='center', 
        color=foreground, fontsize=11, fontweight='bold',
        bbox=dict(boxstyle="round,pad=0.3", facecolor=obsidian, edgecolor=foreground))

# Input/Output
input_box = FancyBboxPatch((0.05, 0.05), 0.25, 0.1,
                          boxstyle="round,pad=0.02",
                          facecolor=taupe,
                          edgecolor=foreground)
ax.add_patch(input_box)
ax.text(0.175, 0.1, 'New Follower\nor Mention', ha='center', va='center', 
        color=obsidian, fontsize=10, fontweight='bold')

output_box = FancyBboxPatch((0.7, 0.05), 0.25, 0.1,
                           boxstyle="round,pad=0.02",
                           facecolor=ttr_yellow,
                           edgecolor=obsidian)
ax.add_patch(output_box)
ax.text(0.825, 0.1, 'Block/Allow\nDecision', ha='center', va='center', 
        color=obsidian, fontsize=10, fontweight='bold')

# Add flow arrows
ax.annotate('', xy=(0.5, 0.75), xytext=(0.3, 0.15),
           arrowprops=dict(arrowstyle='->', color=ttr_yellow, lw=2))
ax.annotate('', xy=(0.5, 0.55), xytext=(0.5, 0.75),
           arrowprops=dict(arrowstyle='->', color=ttr_yellow, lw=2))
ax.annotate('', xy=(0.5, 0.35), xytext=(0.5, 0.55),
           arrowprops=dict(arrowstyle='->', color=ttr_yellow, lw=2))
ax.annotate('', xy=(0.7, 0.1), xytext=(0.5, 0.35),
           arrowprops=dict(arrowstyle='->', color=ttr_yellow, lw=2))

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

plt.tight_layout()
plt.show()
```
